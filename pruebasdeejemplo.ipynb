```python
def recorrer_matriz(matriz):
    numero_de_filas = len(matriz)
    if numero_de_filas == 0:
        print('la matriz esta vacia')
        return 
    
    numero_de_columnas = len(matriz[0])
    print(f'inicializando recorrido en {numero_de_filas} filas x {numero_de_columnas}')
    print('---')

    for i in range(numero_de_filas):
        for j in range(numero_de_columnas):
            elementos = matriz[i][j]
            print(f'elemento en [{i}][{j}]: {elementos}')
            print(f'fila {i} completada')


MAPA = [
    ['.', '.', '.', '.', '.'],
    ['.', '#', '.', '#', '.'],
    ['.', '.', 'P', '.', '.'], # 'P' es la posición inicial del jugador
    ['.', '#', '.', '#', '.'],
    ['.', '.', '.', '.', '.']
]

recorrer_matriz(MAPA)
```
